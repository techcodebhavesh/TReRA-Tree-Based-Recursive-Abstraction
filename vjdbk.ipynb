```python
import os
from groq import Groq
from sentence_transformers import SentenceTransformer
import faiss

# Initialize the embedding model and FAISS index
model = SentenceTransformer("all-MiniLM-L6-v2")
index = faiss.IndexFlatL2(384)  # Assuming 384 dimensions for the embeddings
texts=relevant_contexts

# Populate FAISS index
embeddings = model.encode(texts)
index.add(embeddings)

# Function to retrieve relevant context
def get_relevant_context(query, index, model, texts, top_k=3):
    query_embedding = model.encode([query])
    distances, indices = index.search(query_embedding, top_k)
    return [texts[i] for i in indices[0]]


# Retrieve context for the query
query = "test condtion Maintain Quality Management how to input "
relevant_contexts = get_relevant_context(query, index, model, texts)

# Prepare the prompt for Groq LLM
prompt = f"The user asked: '{query}'\n\nHere is some relevant context:\n{relevant_contexts}\n\nPlease provide an answer based on the above."

# Initialize Groq client
client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

# Generate response using Groq LLM
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="llama-3.3-70b-versatile",
    stream=False,
)

# Output the response
print(chat_completion.choices[0].message.content)
```

```python
import faiss
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient
from qdrant_client.http.models import PointStruct, VectorParams, Distance

# Load SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to create embeddings for text
def create_embedding(text):
    return model.encode([text])[0]

# Initialize Qdrant vector database
def initialize_vector_db():
    client = QdrantClient(path="qdrant_db")  # Local storage

    # Create a collection for storing embeddings
    client.recreate_collection(
        collection_name="documents",
        vectors_config=VectorParams(size=model.get_sentence_embedding_dimension(), distance=Distance.COSINE)
    )

    return client

# Store markdown text and its embedding
def store_document(client, text, doc_id):
    # Compute embedding
    embedding = create_embedding(text)

    # Serialize payload to ensure compatibility
    serialized_text = str(text)

    # Insert text and embedding into Qdrant
    client.upsert(
        collection_name="documents",
        points=[PointStruct(id=doc_id, vector=embedding, payload={"text": serialized_text})]
    )

# Retrieve relevant context based on a query
def retrieve_relevant_context(client, query, top_k=3):
    # Compute embedding for query
    query_embedding = create_embedding(query)

    # Search in Qdrant
    search_results = client.search(
        collection_name="documents",
        query_vector=query_embedding,
        limit=top_k
    )

    return [(result.id, result.payload["text"], result.score) for result in search_results]

# Main script
if __name__ == "__main__":
    # Initialize vector database
    client = initialize_vector_db()

    # Markdown text from pymupdf4llm
    # Ensure only plain text content is processed
    markdown_texts = [str(chunk) for chunk in md_text_words]  # Convert to strings if necessary

    # Store all markdown texts with unique IDs
    for idx, chunk in enumerate(markdown_texts):
        store_document(client, chunk, idx)

    # Example query for retrieval
    query = "Maintain Material Master Defaults"
    results = retrieve_relevant_context(client, query)

    print("Relevant Results:")
    for doc_id, text, score in results:
        print(f"Doc ID: {doc_id}, Score: {score}\n{text}\n")
```
